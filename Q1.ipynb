```python
import numpy as np
import matplotlib.pyplot as plt

def samples_sinusoidal(signal):
    #max points
    mx= (np.max(signal)) #value (the y)
    x_mx= np.where(signal==mx)[0] # the relevats indexs are in the first element of the taple (the x)
    #min points
    mn=np.array(np.min(signal)) #value
    x_mn= np.where(signal==mn)[0] #index

    sample=np.arange(0,100,5) #take the values every 5 "steps"=the indexes of the samples
    indx=sorted(set(sample)|set(x_mn)|set(x_mx)) #combine all the relevant indexes
    final= signal[indx] #creat new array with only the values of the relevant indexes (min&max points with the sample points)
    return final, indx

signal=np.sin(5 * np.pi * 2 *(np.arange(0,1,0.01)))
sample,indx= samples_sinusoidal(signal)
plt.plot(signal,label='original',linestyle='--',color='blue')
plt.plot(indx,sample, label='sample',color='red')
```
